5.
The key difference between method and attributes is that a method will perform an action or operation on the data, whereas an attribute will simply show you the characteristics of the data.

6.
count: Shows value count excluding the missing values if there are any, in every column
mean: The average value of the column
std: The standard deviation of the values in the column, further meaning how deviated the values are from the mean
min: The smallest value in the column
25%, 50%, 75%: Represent the first, second(median), and third quartiles of the column. 
max: The largest value in the column
